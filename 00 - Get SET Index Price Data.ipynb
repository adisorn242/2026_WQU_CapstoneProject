```python
import yfinance as yf
import numpy as np
import pandas as pd
```

# Collect SET Index Data

```python
SET_index = '^SET.BK'

SET_price = df = yf.download(
    tickers=SET_index,
    start="2012-01-01",
    end="2026-01-01",
    interval="1d"
)
```

```python
# Resample SET_price to monthly using the last trading day of each month
# Note if we use monthly data, we will not get full historical data.
SET_price = SET_price.resample('ME').last()
SET_price.index = SET_price.index.to_period('M').to_timestamp()
```
