# Collect SET100 Stock, REITs, IFFs, Gold ETF data (As of 31/12/2025)

```python
# 1. SET100 Symbols
set100_symbols = [
    "AAV", "ADVANC", "AEONTS", "AMATA", "AOT", "AP", "AURA", "AWC", "BA", "BAM",
    "BANPU", "BBL", "BCH", "BCP", "BCPG", "BDMS", "BEM", "BGRIM", "BH", "BJC",
    "BLA", "BTG", "BTS", "CBG", "CCET", "CENTEL", "CHG", "CK", "COM7", "CPALL",
    "CPF", "CPN", "CRC", "DELTA", "DOHOME", "EA", "EGCO", "ERW", "GFPT", "GLOBAL",
    "GPSC", "GULF", "GUNKUL", "HANA", "HMPRO", "ICHI", "IRPC", "IVL", "JAS", "JMART",
    "JMT", "JTS", "KBANK", "KCE", "KKP", "KTB", "KTC", "LH", "M", "MEGA", "MINT",
    "MOSHI", "MTC", "OR", "OSP", "PLANB", "PR9", "PRM", "PTG", "PTT", "PTTEP",
    "PTTGC", "QH", "RATCH", "RCL", "SAWAD", "SCB", "SCC", "SCGP", "SIRI", "SISB",
    "SJWD", "SPALI", "SPRC", "STA", "STECON", "STGT", "TASCO", "TCAP", "TFG",
    "TIDLOR", "TISCO", "TLI", "TOA", "TOP", "TRUE", "TTB", "TU", "VGI", "WHA"
]

# 2. REITs, Infrastructure Funds, and Gold ETF
reit_iff_gld_symbols = [
    "FTREIT", "LHHOTEL", "CPNREIT", "WHART", "AXTRART", "IMPACT", "INETREIT",
    "DIF", "JASIF", "TFFIF", "BTSGIF", "EGATIF", "GLD"
]

# Combine all and add .BK suffix for Yahoo Finance
all_tickers = [s + ".BK" for s in (set100_symbols + reit_iff_gld_symbols)]

# 3. Download Weekly Data (2010 - 2025)
data = yf.download(
    all_tickers,
    start="2010-01-01",
    end="2025-12-31",
    interval="1wk",
    auto_adjust=True,
    progress=False,
    multi_level_index=False
)

# 4. Separate Open and Close into distinct DataFrames
# This discards High, Low, Adj Close, and Volume
df_all_open = data['Open']
df_all_close = data['Close']
```

```python
# Save to CSV files
if True:   # Set as True to export as csv file
  df_all_open.to_csv("SET100_REIT_IFF_GLD_Open_Weekly.csv")
  df_all_close.to_csv("SET100_REIT_IFF_GLD_Close_Weekly.csv")
```
